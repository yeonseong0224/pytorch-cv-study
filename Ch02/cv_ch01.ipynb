#### 1.5 커스텀 클래스 정의하기
개발자가 독자적으로 자신이 클래스를 정의하기도 한다. 이를 '커스텀 클래스를 정의한다'라고 한다.

클래스는 '틀'에 해당하는 개념이고, 인스턴스는 이 '틀'로부터 생성된 개별적인 실체다. 클래스는 '속성'이라고 하는 클래스 안의 변수를 갖는다. 또한, '함수' 또는 '메서드'라고 하는 처리 기능도 존재한다.

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Malgun Gothic'   # 맑은 고딕
mpl.rcParams['axes.unicode_minus'] = False      # 마이너스 깨짐 방지

plt.plot([1,2,3],[1,4,9])
plt.title("한글 그래프 테스트")
plt.xlabel("시간")
plt.ylabel("값")
plt.show()
```

```python
class Point:
    def __init__(self, x, y):  # 이거 두번째 x랑
        self.x = x             # 이거 = x 랑 같음 근데 self.x의 x랑 = x의 x는 다름
        self.y = y
    
    def draw(self):
        plt.plot(self.x, self.y, 'ro', markersize=10)  # 빨간 점으로 표시
```

```python
p1 = Point(2,3)   # 여기서 변수는 인스턴스
p2 = Point(-1, -2)

print(p1.x, p1.y)
print(p2.x, p2.y)
```

```python
p1.draw()
p2.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```

```python
class Circle1(Point):  # 포인트를 상속받는것 circle1은 point의 자식 클래스
    def __init__(self, x, y, radius=1):  # radius는 매개변수로, 기본값은 1로 설정
        super().__init__(x, y)  # super()는 부모 클래스의 메서드를 호출하는 함수 Point의 __init__ 메서드를 호출해서 x와 y를 초기화
        self.radius = radius
    
    def draw(self):
        super().draw()  # 중심점 그리기 -> 빨간점 찍는거
        circle = patches.Circle((self.x, self.y), self.radius, fill=False, edgecolor='blue') # patches는 원을 그릴 수 있게 함
        plt.gca().add_patch(circle)  # 원 추가
```

```python
c1 = Circle1(1, 0, 2)  # 객체 생성
```

```python
# ax = plt.subplot()
p1.draw()
p2.draw()
c1.draw()
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.show()
```

#### 1.6 인스턴스를 함수로 사용하는 방법

클래스에서 생성한 인스턴스를 호출 가능한 함수로 만드는 방법 소개
클래스 정의 안에서 __call__이라는 함수를 정의하고, 이 함수에 원하는 내용을 기술하는 것만으로 충분하다

```python
class H:
    def __call__(self, x):
        print(f"H 객체가 호출되었습니다. x={x}")
        return 2*x**2 + 2


'''Python에서 객체를 함수처럼 호출하면
내부적으로 __call__ 메서드가 실행된다.

즉
f(x) 는 내부적으로
f.__call__(x) 와 같다.

따라서 클래스에 __call__을 정의하면
객체를 함수처럼 사용할 수 있다.
'''
```

```python
x = np.arange(-2, 2.1, 0.25)
print(x)
```

```python
h = H()  # 객체 생성
y = h(x)  # 객체 호출
```

```python
print(y)
```

```python
plt.plot(x, y)
plt.title("H(x) 그래프")
plt.xlabel("x")
plt.ylabel("H(x)")
plt.grid()  # 격자 추가
plt.show()
```
