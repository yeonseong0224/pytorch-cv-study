```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕
mpl.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot([1,2,3],[1,4,9])
plt.title('한글 그래프 테스트')
plt.xlabel('시간')
plt.ylabel('값')
# 마이너스 깨짐 방지
plt.show()
```

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def draw(self):
        plt.plot(self.x, self.y, 'ro' , markersize=10)
```

```python
p1 = Point(2,3)
p2 = Point(-1, -2)

print(p1.x, p1.y)
print(p2.x, p2.y)
```

```python
p1.draw()
p2.draw()
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

```python
import matplotlib.patches as patches
```

```python
class Circle1(Point):
    def __init__(self, x, y, radius):
        super().__init__(x, y)
        self.radius = radius
        
    def draw(self):
        super().draw()
        circle = patches.Circle((self.x, self.y),self.radius, fill=False, edgecolor='blue')
        plt.gca().add_patch(circle)
```

```python
c1 = Circle1(1,0,2)
```

```python
p1.draw()
p2.draw()
c1.draw()
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

```python
class H:
    def __call__(self, x):
        print(f"H 객체가 호출되었습니다. x={x}")
        return 2*x**2+2
    
```

```python
import numpy as np
```

```python
x = np.arange(-2,2.1,0.25)
print(x)
```

```python
h = H()
y = h(x)
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.title('H 함수 그래프')
plt.xlabel('x')
plt.ylabel('H(x)')
plt.grid()
plt.show()
```
